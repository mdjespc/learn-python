```python
filename = "ascii-codes.txt"
#Create dictionary
D = dict()
f = open(filename, "r")
#Read each line in file
for x in f:
  x = x.split(',')
  value = x[0] #Binary String
  key = x[1].replace("\n", "") #Character
  D[key] = value
f.close()

print(D)
```

```python
'''
Creates a new file where every character in a given sentence is translated
into its ASCII code
'''
def ASCII_translation(sentence, dictionary, filename):
  #Open the new file
  try:
    newFile = open(filename, "w")
    print('\'%s\' has been created' %filename)
  except:
    print('Could not create file')
  #Read through every character inside the sentence
  for char in sentence:
    key = char
    #Handle both exceptions (space and comma)
    if char == " ":
      key = 'space'
    if char == ",":
      key = 'comma'
    #Is the character in the dictionary?
    if key in dictionary.keys():
      newFile.write('%s %s\n' %(char, dictionary[key]))
    else:
      newFile.write('%s UNKNOWN\n' %key)

  newFile.close()
```

```python
string1 = 'Buck lived at a big house in the sun-kissed Santa Clara Valley. Judge Millers place, it was called!'
string2 = 'Bozeman, MT  59717'
string3 = 'The value is ~$25.00'
ASCII_translation(string1, D, 'output-1.txt')
ASCII_translation(string2, D, 'output-2.txt')
ASCII_translation(string3, D, 'output-3.txt')
```

```python
class Person:
  #Defining the properties of a person
  age = 0
  name = ""
  #Constructor
  def __init__(self, age, name):
    self.age = age
    self.name = name

  '''
  METHODS --- obj.method(parameters)
  '''
  #Setters -- Modify an attribute of an object
  def setAge(self):
    self.age = int(input())
  def setName(self):
    self.name = input()
  #Getters -- Return an attribute
  def getAge(self):
    return self.age
  def getName(self):
    return self.name
  #Functionality methods
  def display(self):
    print('{Name: %s\tAge: %d}'%(self.name, self.age))
```

```python
#Initialize an object
A = Person(20, 'Miguel')
B = Person(33, 'Matthew')
#Use our display method
A.display()
B.display()
var = B.getName()
var
```
