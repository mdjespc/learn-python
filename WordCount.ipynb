```python
import string
```

```python
file = 'raven.txt'
text = open(file, 'r')
#Make the dictionary
wordTable = dict()

#Reading each line in the text
for i in text:
  i = i.lower()
  #String that contains the cleaned up version of the line
  line=''
  for j in i:
    if j in string.ascii_lowercase or j == ' ':
      line += j
  print(line)
  line = line.split()
  for word in line:
    if word not in wordTable.keys():
      wordTable[word] = 1
    else:
      wordTable[word] += 1
```

```python
maxValue = float('-inf')
maxKey = ''
for kv in wordTable:
  print(kv, ':', wordTable[kv])
  if wordTable[kv] > maxValue:
    maxValue = wordTable[kv]
    maxKey = kv
text.close()
```

```python
print('The most common word is \'%s\' with %d appearances' %(maxKey, maxValue))
```
