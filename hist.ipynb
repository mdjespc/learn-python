```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Initialize dataframe
df = pd.read_csv('US-Gun-Violence.csv')
df.nunique()
```

```python
#Initialize figure
fig = plt.figure(num = 'CSCI 127', figsize=[18, 6])
plt.title('US Gun Violence Records per State')
#Plot a histogram that shows the number of gun violence incidents per state
plt.hist(df['state'], bins = 48, color = 'orange', histtype= 'stepfilled')
plt.grid(True)
#Labels
plt.xlabel('US State')
ax = fig.get_axes()[0]
plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.ylabel('Gun Violence Incidents')
plt.show()
```
